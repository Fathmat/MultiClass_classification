```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import datetime
```

```python
df = pd.read_csv('/content/AppleTEST.csv')
#df['meta_publish_date'] = pd.to_datetime(df['meta_publish_date'])
df.head()
```

```python
AAPL_df= yf.download(tickers='AAPL')
```

```python
AAPL_df_STR= AAPL_df.loc['2010-01-01':,:]
if isinstance(AAPL_df_STR.columns, pd.MultiIndex):
    AAPL_df_STR.columns = AAPL_df_STR.columns.droplevel(1)
AAPL_df_STR = AAPL_df_STR.reset_index()
#AAPL_df_STR['date'] = pd.to_datetime(AAPL_df_STR.index)
AAPL_df_STR.head()
```

```python
AAPL_df_STR['Date'] = pd.to_datetime(AAPL_df_STR['Date'])
AAPL_df_STR.set_index('Date', inplace=True)
AAPL_df_STR.head()
```

```python
AAPL_df_STR.index = pd.to_datetime(AAPL_df_STR.index)
```

```python
sentiment_data= df['meta_publish_date'].copy()
sentiment_data = pd.DataFrame(sentiment_data)
sentiment_data.head()
```

```python
sentiment_data['Sentiment'] = df['Sentiment_from_the report']
sentiment_data['Sentiment1'] = df['Overall_sentiment_from the event']
sentiment_data['Sentiment2'] = df['Share_Repurchase_Signal']
sentiment_data.rename(columns={'meta_publish_date':'Date'}, inplace=True)
sentiment_data.head()
```

```python
sentiment_data['Date'] = pd.to_datetime(sentiment_data['Date'])
```

```python
# Remove timezone from the 'Date' column in AAPL_df_STR (if it has a timezone)
AAPL_df_STR.index = AAPL_df_STR.index.tz_localize(None)

# Now, you can merge the two DataFrames
merged_df = pd.merge(AAPL_df_STR, sentiment_data, left_index=True, right_on='Date', how='left')

merged_df.head()
```

```python
merged_df['Sentiment'].fillna(0, inplace=True)
merged_df['Sentiment1'].fillna(0, inplace=True)
merged_df['Sentiment2'].fillna(0, inplace=True)
merged_df.head()
```

```python
data_df= merged_df['Date']
data_df = pd.DataFrame(data_df)
data_df['Sentiment'] = merged_df['Sentiment']
data_df['Sentiment1'] = merged_df['Sentiment1']
data_df['Sentiment2'] = merged_df['Sentiment2']
data_df['Close'] = merged_df['Close']
data_df['Open'] = merged_df['Open']
data_df['High'] = merged_df['High']
data_df['Low'] = merged_df['Low']
data_df['Volume'] = merged_df['Volume']
data_df.set_index('Date', inplace=True)
data_df.head()
```

```python
merged_df['Sentiment'] = merged_df['Sentiment'].astype(int)
merged_df['Sentiment1'] = merged_df['Sentiment1'].astype(int)
merged_df['Sentiment2'] = merged_df['Sentiment2'].astype(int)
merged_df.head()
```

```python
from backtesting import Backtest, Strategy
```

```python
class OverallSentimentStrategy(Strategy):
    already_bought = False

    def init(self):
        pass

    def next(self):
        sentiment = self.data.Sentiment[-1]

        if sentiment > 0 and not self.already_bought:
            self.buy()
            self.already_bought = True
        elif sentiment < 0 and self.already_bought:

            self.sell()
            self.already_bought = False
        else:
            pass

# Backtest
bt = Backtest(data = data_df, strategy=OverallSentimentStrategy, cash=10000, commission=.002)

output = bt.run()
output
```

```python
bt.plot()
```
